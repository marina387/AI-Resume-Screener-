## Мини-теория

### Распределение
Показывает, как часто встречаются значения в данных.

### Нормальное распределение
- Симметричное
- Среднее ≈ медиана
- 68% значений лежат в диапазоне ±1 std

### Гистограмма
Строится через matplotlib:
- plt.hist()
- plt.show()

В конце: Run all → все assert должны пройти.


## 1/10 Генерация нормального распределения

```python
import numpy as np

# Генерируем 1000 случайных чисел, которые следуют нормальному распределению
data = np.random.normal(0, 1, 1000)

# Проверяем количество сгенерированных чисел и выводим результат
print(len(data))
```

## 2/10 Среднее и стандартное отклонение

```python
mean = np.mean(data) # среднее значение
std = np.std(data) # стандартное отклонение
mean, std
```

## 3/10 Гистограмма нормального распределения

```python
import matplotlib.pyplot as plt  # библиотека для графиков

plt.hist(data, bins=30)        # строим гистограмму из данных с 30 интервалами
plt.title("Normal Distribution") # заголовок графика
plt.show()                     # показываем график
```

## 4/10 Добавление выброса

```python
data_out = np.append(data, 10)  # добавляем число 10 в конец массива data
len(data_out)                    # выводим длину нового массива (1001)
```

## 5/10 Гистограмма с выбросом

```python
plt.hist(data_out, bins=30)      # гистограмма данных с выбросом (значение 10)
plt.title("With Outlier")        # заголовок указывает на наличие выброса
plt.show()                       # отображаем график
```

## 6/10 Медиана

```python
median = np.median(data_out)  # вычисляем медиану (устойчива к выбросам)
median                        # выводим значение медианы
```

## 7/10 Проверка правила 68%

```python
lower = mean - std                    # нижняя граница интервала (среднее - 1σ)
upper = mean + std                    # верхняя граница интервала (среднее + 1σ)
within = np.sum((data >= lower) & (data <= upper))  # считаем точки внутри интервала
within / len(data)                     # доля точек в интервале (≈68% для норм. распр.)
```

## 8/10 Равномерное распределение

```python
uniform = np.random.uniform(-3, 3, 1000)  # 1000 чисел из равномерного распределения от -3 до 3
plt.hist(uniform, bins=30)                # строим гистограмму с 30 интервалами
plt.title("Uniform Distribution")         # заголовок "Равномерное распределение"
plt.show()                                # отображаем график
```

## 9/10 Сравнение средних

```python
np.mean(data), np.mean(uniform)  # средние нормального и равномерного распределений (оба ≈ 0)
```

```python
assert len(data) == 1000                        # проверка: в data ровно 1000 элементов
assert isinstance(mean, float)                  # проверка: mean имеет тип float
assert isinstance(std, float)                   # проверка: std имеет тип float
print("✅ BLOCK04 LESSON02: all tests passed")  # сообщение об успешном прохождении тестов
```
